```python
from sympy.logic.boolalg import *
from sympy.abc import A, B, C, D, E, F, G
from sympy.logic.inference import pl_true
import random
from sympy.logic.inference import literal_symbol
from sympy.logic.inference import satisfiable
```

```python
def get_random_assign(variables) :
    ass = {}
    for var in variables :
        if (random.random() <= 0.5) :
            ass[var] = 0
        else :
            ass[var] = 1
    
    return ass
```

```python
def sat_random(cnf, variables) :
    n = len(variables)
    
    #for (i = 1, 2,..., ⌈(4/3)n⌉) { // n is the number of variables in ϕ
    for i in range(1, round(((4/3)**(n))+1)) :
        
        #Randomly choose an assignment β ∈ {0, 1}n under the uniform distribution;
        random_assign = get_random_assign(variables)
        
        for i in range(1, n) :
            #if (ϕ(β) = 1) return the satisfying assignment β of ϕ and halt;
            if (pl_true(cnf, random_assign) == True) :
                
                return random_assign
            
            else :
                #        Choose any clause C = (x ∨ y ∨ z) with C(β)=0;
                for j in range(0, len(cnf.args)) :
                    if (pl_true(cnf.args[j], random_assign) == False) :
                        false_clause = cnf.args[j]
                        
                #       Randomly choose a literal ℓ ∈ {x, y, z} under the uniform distribution;
                
                literal = false_clause.args[random.randint(0, 2)]
                
                #      Determine the bit βℓ ∈ {0, 1} in β that assigns ℓ;
                #      Modify βℓ to 1 − βℓ in β;
                random_assign[literal_symbol(literal)] = 1 - random_assign[literal_symbol(literal)]

    return "NOT SATISFIABLE"
```
